# Example API calls to overpass api

```python
import requests

query = """
[out:json];
(
  way["highway"](48.1,11.5,48.2,11.6);
  way["building"](48.1,11.5,48.2,11.6);
  node["amenity"="waste_basket"](48.1,11.5,48.2,11.6);
  node["amenity"="waste_container"](48.1,11.5,48.2,11.6);
);
out center;
"""

response = requests.get("https://overpass-api.de/api/interpreter", params={'data': query})
data = response.json()
```

```python
data
```

```python
query_waste = """
[out:json];
(
    node["amenity"="waste_basket"](48.1,11.5,48.2,11.6);
    node["amenity"="waste_container"](48.1,11.5,48.2,11.6);
);
out center;
"""

response_waste = requests.get("https://overpass-api.de/api/interpreter", params={'data': query_waste})
data_waste = response_waste.json()
```
